```r
p_size <- seq(20, 230, 1)
p_size
```

```r
glider <- read.table('../reg2020/sugar_glider.txt',header=T)
logit_m2 <- glm(occurr ~ p_size_km, family=binomial(link=logit), data=glider)
```

```r
hat_eta <- predict(logit_m2, list(p_size_km=p_size),type='link')
```

```r
par(mfrow=c(1,2))
attach(glider)
plot(p_size_km, occurr, xlab='size of range(x),', ylab='hat pi(x) \\ occurr', sub='(a)', pch=20)
lines(p_size, exp(hat_eta)/(1+exp(hat_eta)), lwd=1.5, col='red')

#   
glider_g <- read.csv('../reg2020/sugar_glider_binomial_g.csv')

plot(glider_g$p_size_med, 
    glider_g$cases/glider_g$count, 
    xlab='size of range(x)', 
    ylim=c(0,1), 
    ylab= 'hat pi(x) \\ sample prop',
    sub='(b)',
    pch=20,
    col='blue')
lines(p_size, exp(hat_eta)/(1+exp(hat_eta)), lwd=1.5, col='red')
```
