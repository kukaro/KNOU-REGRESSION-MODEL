```r
glider <- read.table('../reg2020/sugar_glider.txt',header=T)
t(glider)
```

```r
logit_ml <- glm(occurr~p_size_km+con_metric,family=binomial(link=logit),data=glider)
```

```r
logit_ml
```

```r
logit_ml.summary <- summary(logit_ml)
logit_ml.summary
#Null deviance -> 상수항만 사용 했을때의 이탈도
#Residual deviance -> 설명변수를 포함시켰을 때의 이탈도
```

```r
logit_ml.summary$coefficients
```

```r
#모형의이 유의성 검정
library(imager)
```

```r
#모형의 유의성 검정
plot(load.image('./img/001.png'), axes = F)
```

```r
residual.deviance <- logit_ml.summary$deviance
null.deviance <- logit_ml.summary$null.deviance
df.residual <- logit_ml.summary$df.residual
df.null <- logit_ml.summary$df.null
p.value <- 1 - pchisq(null.deviance - residual.deviance, df.null - df.residual)
p.value
# p값이 매우 작으면 대립가설의 모형이 유의하다
```

```r
residual.deviance
null.deviance
df.residual
df.null
```

```r
logit_ml.anova <- anova(logit_ml)
logit_ml.anova
```

```r
logit_m0 <- glm(occurr~1, family=binomial(link=logit),data=glider)
logit_m0
```

```r
logit_ml
```

```r
logit_m0$coefficients
logit_ml$coefficients
```

```r
logit_m0.anova <- anova(logit_m0, logit_ml, test='Chisq')
logit_m0.anova
#구해진 p.value가 수작업으로 구한것과 같다
```

```r
# 모형의 적합성 검정
plot(load.image('./img/002.png'), axes = F)
```

```r
residual.deviance <- logit_ml.summary$deviance
df.residual <- logit_ml.summary$df.residual
p.value <- 1 - pchisq(residual.deviance, df.residual)
p.value
# p.value가 0.05 이상이므로 귀무가설을 기각하지못함.
# 모형이 적합하다고 판단
```

```r
# 모형의 선택
# 변수 선택 : x1이 선택 된 모형에서 x2 를 추가하는 것이 적절한 지 검정
plot(load.image('./img/003.png'), axes = F)
```

```r
logit_m2 <- glm(occurr ~ p_size_km, family=binomial(link=logit), data=glider)
logit_m1 <- glm(occurr ~ p_size_km + con_metric, family=binomial(link=logit), data=glider)
logit_m2
'######################################################################'
logit_m1
```

```r
anova(logit_m2, logit_m1, test='Chisq')
```

```r
# p.value가 0.05보다 크므로 둘은 유의미한 차이가 있다고 본다.
```

```r
AIC(logit_m2, logit_m1)
#AIC가 더 작은 m2모형을 채택한다
```

```r
#모형의 선택 : 변수선택방법 이용
library(MASS)
stepAIC(logit_m1, direction='both')
```
