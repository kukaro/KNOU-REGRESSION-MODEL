```r
health <- read.table('../reg2020/health.txt', header = T)
health <- health[, -1]
```

```r
t(health)
```

```r
# Y와 X4의 관계가 가장 크고 X1이 그 다음임을 확인할 수 있다.
cor(health)
```

```r
h1.lm <- lm(Y ~ X4, data = health)
h1.lm.summary <- summary(h1.lm)
h1.lm.summary
```

```r
h2.lm <- lm(Y ~ X4 + X1, data = health)
h2.lm.summary <- summary(h2.lm)
h2.lm.summary
```

```r
h3.lm <- lm(Y ~ X4 + X1 + X2, data = health)
h3.lm.summary <- summary(h3.lm)
h3.lm.summary
```

```r
h4.lm <- lm(Y ~ X4 + X1 + X2 + X3, data = health)
h4.lm.summary <- summary(h4.lm)
h4.lm.summary
```

```r
anova(h1.lm, h2.lm)
```

```r
h0.lm <- lm(Y ~ X1, data = health)
h0.lm.summary <- summary(h0.lm)
h0.lm.summary
```

```r
anova(h0.lm, h2.lm)
```

```r
library(car)
avPlots(h2.lm)
```

```r
health.lm.anova <- anova(h2.lm)
health.lm.anova
```

```r
plot(health$X1, as.vector(h2.lm$residuals))
```

```r
plot(health$X4, as.vector(h2.lm$residuals))
```
