```r
# 몸무게
X1 <- c(217, 141, 152, 153, 180,
        193, 162, 180, 205, 168,
        232, 146, 173, 155, 212)
```

```r
# 분당 정지 맥박수
X2 <- c(67, 52, 58, 56, 66,
        71, 65, 80, 77, 74,
        65, 68, 51, 64, 66)
```

```r
# 근력
X3 <- c(260, 190, 203, 183, 170,
        178, 160, 170, 188, 170,
        220, 158, 243, 198, 220)
```

```r
# 1.4마일 시험주행속도
X4 <- c(91, 66, 68, 70, 77,
        82, 74, 84, 83, 79,
        72, 68, 56, 59, 77)
```

```r
# 1마일 주행속도
Y <- c(481, 292, 338, 357, 396,
       429, 345, 469, 425, 358,
       393, 346, 279, 311, 401)
```

```r
df <- data.frame(Y, X1, X2, X3, X4)
```

```r
t(df)
```

```r
# X4가 종속변수와의 관계가 큰것을 확인할 수 있다.
cor(df)
```

```r
df.lm <- lm(Y ~ X1 + X2 + X3 + X4, data = df)
df.lm
```

```r
#변수들 중 X4가 유의수준 0.05보다 낮으므로 귀무가설을 기각한다. 즉 X4는 유의미한 영향을 미친다고 할 수 있다.
df.lm.summary <- summary(df.lm)
df.lm.summary
```

```r
# 추가변수그림을 봤을 때 X4가 가장 적합도가 높아보임을 확인할 수 있다.
library(car)
avPlots(df.lm)
```

```r
df.lm.anova <- anova(df.lm)
df.lm.anova
```
