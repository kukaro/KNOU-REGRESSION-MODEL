```r
market2 <- read.table('../reg2020/market-2.txt', header=T)
```

```r
t(market2)
```

```r
X <- market2[,c(2,3)]
t(X)
```

```r
X <- cbind(1,X)
t(X)
```

```r
Y <- market2[,4]
t(Y)
```

```r
X <- as.matrix(X)
t(X)
```

```r
Y <- as.matrix(Y)
t(Y)
```

```r
XTX <- t(X) %*% X
XTX
```

```r
XTXI <- solve(XTX)
XTXI
```

```r
XTY <- t(X) %*% Y
XTY
```

```r
beta <- XTXI %*% XTY
beta
```

```r
beta <- round(beta, 3)
beta
```
