```r
glider <- read.csv('../reg2020/sugar_glider_binomial.csv')
```

```r
attach(glider)
```

```r
probit.lm <- glm(occurr~p_size_km, family=binomial(link=probit))
```

```r
probit.lm
```

```r
probit.lm.summary <- summary(probit.lm)
probit.lm.summary
```

```r
probit.lm.summary$coefficients
```

```r
predict(probit.lm, list(p_size_km=150), type='response')
```

```r
plot(p_size_km, occurr, type='n', xlab='size of range(x)', ylab='hat pi(x) \\ occurr')
rug(jitter(p_size_km[occurr == 0]))
rug(jitter(p_size_km[occurr == 1]))

logit.m2 <- glm(occurr~p_size_km, family=binomial(link=logit),data=glider)

# 설명변수 구간 나누기 및 각 구간에서 표본비율 구하기
cl.intr <- cut(p_size_km, 5)
s.prop <- as.vector(tapply(occurr, cl.intr, mean))
md.size <- as.vector(tapply(p_size_km, cl.intr, median))

x <- seq(23, 226, 1)
hat.pi.p <- predict(probit.lm, list(p_size_km=x), type='response')
lines(x, hat.pi.p, col='blue', lty=1, lwd=1.5)

hat.pi.l <- predict(logit.m2, list(p_size_km=x), type='response')
lines(x, hat.pi.l, col='red', lty=2, lwd=1.5)

lines(md.size, s.prop, type='p', pch=20, cex=1.5, col='blue')
legend('topleft', c('probit model','logistic modle'), lty=c(1,2), col=c('blue','red'), cex=0.7)
```
